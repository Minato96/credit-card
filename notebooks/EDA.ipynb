## Feature Analysis
based on the column names we can easily guess that V1-V28 are PCA transformed. So we dont need to process them nor scale them further as they are most likely already in perfect state

```python
fig, axes = plt.subplots(1, 2, figsize=(18, 4))

sns.histplot(df['Amount'], ax=axes[0], color='r', bins=50, kde=True)
axes[0].set_title('Distribution of Transaction Amount')

sns.histplot(df['Time'], ax=axes[1], color='b', bins=50, kde=True)
axes[1].set_title('Distribution of Transaction Time')

plt.show()
```
